1. Coût marginal et coût moyen.

On les fonctions de coût total,  moyen, et marginal suivantes:
\begin{align*}
 CT(q) = q^3 - 4q^2 + 8q &\Rightarrow C_M(q)\coloneqq \frac{CT(q)}{q} = q^2 - 4q + 8\\
 &\Rightarrow  C_m(q)\coloneqq  \frac{\partial CT}{\partial q}(q) = 3q^2 - 8q + 8,
\end{align*}

et la demande inverse:

\begin{align*}
p(q)&= -2q + 8.
\end{align*}


```python
def CM_func(q):
    return q**2 - 4*q + 8

def Cm_func(q):
    return 3*q**2 - 8*q + 8

def Dmde_func(q):
    return -2 * q + 8
```

```python
import numpy as np
```

```python
q_max = 15
N = 1000
q_vec = np.linspace(0, q_max ,  N)
# Évaluation des fonctions en utilisant les opérations sur les array.
CM_vec = CM_func(q_vec)
Cm_vec = Cm_func(q_vec)
Dmde_vec = Dmde_func(q_vec)

# Alternative avec boucle.
#CM_vec = np.empty(N)
#Cm_vec = np.empty(N)
#Dmde_vec = np.empty(N)
#for i, q in enumerate(q_vec):
#    CM_vec[i] = CM_func(q)
#    Cm_vec[i] = CM_func(q)
#    Dmde_vec[i] = Dmde_func(q)
```
